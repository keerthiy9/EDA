**Data Transformations**
- Generally data has 3 types of distributions
    - i) Normal distribution
    - ii) Posotive skewed
    - iii) Negative skewed
- All the maths concepts developed by assumption as Data follows Normal distribution
- Data skew happens beacause of many reasons, in that one reason is outliers
- Once we perform the outlier analysis then we can apply transformation methods , if the data does not follows normality
- We have some transformatioon methods mentioned below
    - Log transformation
    - Exponential transformation
    - Reciprocal transformation
    - Square root transformation
    - Power transformation

**Reciprocal Transformations**

- Reciprocal will fail when values has very small or low
- For example x=0 then 1/0 fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data )
plt.hist(reciprocal_data,bins=20,label='Reciprocal_data')
plt.show()
```

**Square root transformation**

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,4,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,4,2).hist(exp_data,bins=20,label='Log')
plt.legend()
plt.subplot(1,4,3).hist(reciprocal_data,bins=20,label='Reciprocal')
plt.legend()
plt.subplot(1,4,4).hist(square_root_data,bins=20,label='Square')
plt.legend()
plt.show()
```
