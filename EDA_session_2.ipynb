**create data frame using dictionary**

```python
name=['akash','aadhya','arya']
age=[25,29,31]
pd.DataFrame(zip(name,age),columns=['names','age'])
```

```python
dict1={'names':['akash','aadhya','arya'],
       'age': [25,29,31]}
dict1
```

```python
pd.DataFrame(dict1)
# keys become columns 
# values become rows
# no need to provide column name seperately
```

```python
dict2={'names':'arya',
       'age': 25,
      'city':'hyd'}
pd.DataFrame(dict2)
```

```python
dict2={'names':'arya',
       'age': 25,
      'city':'hyd'}
pd.DataFrame(dict2,index=[1,2,3])
```

```python
dict2={'names':'arya',
       'age': 25,
      'city':'hyd'}
pd.DataFrame(dict2,index=[1,2])
```

```python
dict2={'names':'arya',
       'age': 25,
      'city':'hyd'}
pd.DataFrame(dict2,index=[1])
```
